```python
#data handling/modeling
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import scipy.stats as stats

# visualization
%matplotlib inline
import mpld3
mpld3.enable_notebook()
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
from pycountry_convert import country_alpha2_to_country_name, country_name_to_country_alpha3
import os
```

```python
pd.set_option('display.max_columns', None)
ht = pd.read_csv("The Global Dataset 14 Apr 2020.csv", na_values='-99',low_memory=False)
df = ht.drop(['By using this data you agree to the Terms of Use: https://www.ctdatacollaborative.org/terms-use'], axis=1)
df = pd.DataFrame(df)
df.head()
```

```python
def get_alpha3(col):
    try: iso_3 =  country_name_to_country_alpha3(col)
    except: iso_3 = 'Unknown'
    return iso_3

def get_name(col):
    try: name =  country_alpha2_to_country_name(col)
    except: name = 'Unknown'
    return name
```

```python
df['country'] = df['citizenship'].apply(lambda x: get_name(x))
df['exploitCountry'] = df['CountryOfExploitation'].apply(lambda x: get_name(x))
```

```python
df.head()
```

```python
df.fillna('None', inplace=True)
```

```python
df_gender = pd.get_dummies(df.gender).astype(int)
df_gender = df_gender[['Female']]
df_status = pd.get_dummies(df.majorityStatusAtExploit).astype(int)
df_status = df_status[['Adult','Minor']]
df_control = df.meansOfControlConcatenated.str.get_dummies(sep = ";")
df_typeExploit = df.typeOfExploitConcatenated.str.get_dummies(sep = ";")
df_labour = df.typeOfLabourConcatenated.str.get_dummies(sep = ";")
df_sexLabour = df.typeOfSexConcatenated.str.get_dummies(sep = ";")
df_relationship = df.RecruiterRelationship.str.get_dummies(sep = "; ")
```

```python
df.columns
```

```python
df = df.drop(columns=['Datasource', 'gender',
       'majorityStatus', 'majorityStatusAtExploit', 'majorityEntry',
       'citizenship', 'meansOfControlDebtBondage',
       'meansOfControlTakesEarnings', 'meansOfControlRestrictsFinancialAccess',
       'meansOfControlThreats', 'meansOfControlPsychologicalAbuse',
       'meansOfControlPhysicalAbuse', 'meansOfControlSexualAbuse',
       'meansOfControlFalsePromises', 'meansOfControlPsychoactiveSubstances',
       'meansOfControlRestrictsMovement', 'meansOfControlRestrictsMedicalCare',
       'meansOfControlExcessiveWorkingHours', 'meansOfControlUsesChildren',
       'meansOfControlThreatOfLawEnforcement',
       'meansOfControlWithholdsNecessities',
       'meansOfControlWithholdsDocuments', 'meansOfControlOther',
       'meansOfControlNotSpecified', 'meansOfControlConcatenated',
       'isForcedLabour', 'isSexualExploit', 'isOtherExploit', 'isSexAndLabour',
       'isForcedMarriage', 'isForcedMilitary', 'isOrganRemoval',
       'isSlaveryAndPractices', 'typeOfExploitConcatenated',
       'typeOfLabourAgriculture', 'typeOfLabourAquafarming',
       'typeOfLabourBegging', 'typeOfLabourConstruction',
       'typeOfLabourDomesticWork', 'typeOfLabourHospitality',
       'typeOfLabourIllicitActivities', 'typeOfLabourManufacturing',
       'typeOfLabourMiningOrDrilling', 'typeOfLabourPeddling',
       'typeOfLabourTransportation', 'typeOfLabourOther',
       'typeOfLabourNotSpecified', 'typeOfLabourConcatenated',
       'typeOfSexProstitution', 'typeOfSexPornography',
       'typeOfSexRemoteInteractiveServices', 'typeOfSexPrivateSexualServices',
       'typeOfSexConcatenated', 'isAbduction', 'RecruiterRelationship',
       'CountryOfExploitation', 'recruiterRelationIntimatePartner',
       'recruiterRelationFriend', 'recruiterRelationFamily',
       'recruiterRelationOther', 'recruiterRelationUnknown'])
```

```python
# All data cleaned
df_cleaned = df.merge(df_gender,left_index=True,right_index=True)
df_cleaned = df_cleaned.merge(df_status,left_index=True,right_index=True)
df_cleaned = df_cleaned.merge(df_control,left_index=True,right_index=True)
df_cleaned = df_cleaned.merge(df_typeExploit,left_index=True,right_index=True)
df_cleaned = df_cleaned.merge(df_labour,left_index=True,right_index=True)
df_cleaned = df_cleaned.merge(df_sexLabour,left_index=True,right_index=True)
df_cleaned = df_cleaned.merge(df_relationship,left_index=True,right_index=True)
```

```python
# type of labour cleaned
df_labour_clean = df.merge(df_gender,left_index=True,right_index=True)
df_labour_clean = df_labour_clean.merge(df_status,left_index=True,right_index=True)
df_labour_clean = df_labour_clean.merge(df_labour,left_index=True,right_index=True)
```

```python
# type of sexLabour cleaned
df_sexLabour_clean = df.merge(df_gender,left_index=True,right_index=True)
df_sexLabour_clean = df_sexLabour_clean.merge(df_status,left_index=True,right_index=True)
df_sexLabour_clean = df_sexLabour_clean.merge(df_sexLabour,left_index=True,right_index=True)
```

```python
# type of relationship cleaned
df_relationship_clean = df.merge(df_gender,left_index=True,right_index=True)
df_relationship_clean = df_relationship_clean.merge(df_status,left_index=True,right_index=True)
df_relationship_clean = df_relationship_clean.merge(df_relationship,left_index=True,right_index=True)
```
